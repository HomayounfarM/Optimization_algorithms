## Two simple optimization examples using scipy.minimize
# 1- min f(x)

```python
# import libraries
import scipy
import numpy as np
import scipy.optimize as spo
import matplotlib.pyplot as plt
```

```python
# Function to minimize

def f(x):
    x = np.array(x)
    y = x**2 - 12*x + 20
    return y
```

```python
# initial values 
x_ini = 2.0
```

```python
# Optimizing function
result = spo.minimize(f, x_ini, options = {"disp" : True})
```

```python
# print results
if result.success:
    print("Success")
    print(f"x = {result.x} y = {result.fun})")
else:
    print("Sorry, could not find a minimum.")    
```

```python
type(result.x)
```

```python
# plot the function
xx = np.linspace(-1,12,100)
yy = f(xx)
plt.plot(xx,yy)
x = np.array(result.x)
y = np.array(result.fun)
plt.plot(x,y, marker="o", markersize=10, markeredgecolor="red")
```

# 2- max f(x,y)

```python
# function to maximize

def ff(xy):
    x = np.array(xy[0])
    y = np.array(xy[1])
    area = x * y
    return -area
```

```python
# initial values 
xy_ini = [50,50]
```

```python
# Bounds and constraints
bnds = ((1, 100), (1, 100))

cons = ({'type' : 'eq', 'fun': lambda xy: (2*xy[0]) + xy[1] -100})
```

```python
# Optimizing function
result = spo.minimize(f, xy_ini, options = {"disp" : True}, constraints = cons, bounds = bnds)
```

```python
# print results
if result.success:
    print("Success")
    xy = result.x
    x = xy[0]
    y = xy[1]
    print(f"x = {x} y = {y})")
else:
    print("Sorry, could not find a minimum.")    
```

```python
# plot the function
xx = np.linspace(0, 100, 101)
yy = np.linspace(0, 100, 101)

XX, YY = np.meshgrid(xx, yy)
ZZ = ff([XX, YY])

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(XX, YY, ZZ, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(x,y,ff([x,y]), marker="*")
```
