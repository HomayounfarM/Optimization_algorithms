##Gradient descent:
Gradient descent is a fundamental optimization algorithm in machine learning used to minimize a loss function, which measures how well a model's predictions match the actual data. Here is an example using this algorithm to optimize a quadratic function:

```python
# Install libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  A quadratic function
def f(x):
    x = np.array(x)
    y = (x**2 - 12*x + 20)
    return y
# First derivative of the quadratic function
def f_prim(x):
    x = np.array(x)
    y = (2*x - 12)
    return y

# parameters of the gradient descent algorithm
a = 0.1                # Learning rate
e = 0.0001             # threshold error
x = np.linspace(-10,20)
x1 = np.linspace(1, 5, num=50)
y = f(x)
w = np.array([0])
f_w = np.array([0])
w[0] = 0
diff = 1
i = 0

# Optimizing function
while diff > e:
    w = np.append(w, w[i]-a*f_prim(w[i]))
    f_w = np.append(f_w, f(w[i+1]))
    diff = w[i+1]-w[i]
    i+=1
print('*'*50)
print(e)
print(diff)
print(f_w)
print(w)
print('*'*50)

# ploting the f(w) function
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x,f(x))
ax2.plot(np.linspace(1, len(f_w[1:]),len(f_w[1:])), f_w[1:])
```
